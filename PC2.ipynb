Generamos nuestra tabla de datos para poder visualizar el proceso de limpieza y formateo.

```python
indice_desarrolloLink="https://github.com/LuilliVari/PC2/raw/refs/heads/main/HDI.xlsx"
#Llamamos a la libreria pandas para elaborar el DataFrame.
import pandas as pd
indice_desarrollo=pd.read_excel(indice_desarrolloLink)
#Imprimimos el DataFrame
indice_desarrollo
```

```python
#Empezamos la limpieza eliminando los espacios entre las palabras de los indices.
#Para ello llamamos las columnas del DataFrame.
indice_desarrollo.columns=indice_desarrollo.columns.str.replace(r'\s','',regex=True)
#Volvemos a imprimir el DataFrame para actualizarlo.
indice_desarrollo
```

Se detecta una cantidad de filas que no poseen datos abrumando la comprension y sintesis de los datos. Tambien hay un claro desorden en los indices con algunas columnas con datos inservibles o en blanco.

```python
#Usamos la funcion ".drop" para eliminar las columnas que sostienen a los datos innecesarios o en blanco.
toDrop=["Back","Unnamed:3","Unnamed:5","Unnamed:7","Unnamed:9","Unnamed:11","Unnamed:12","Unnamed:13","Unnamed:14"]
indice_desarrollo.drop(columns=toDrop,inplace=True)
#Se actualiza el DataFrame.
indice_desarrollo
```

```python
#Modificamos los indices para empezar a ordenar las columnas.
#Para ello asignamos una nueva variable para renombrar los indices.
newIndiceDesarrolloNames={'Table1.HumanDevelopmentIndexanditscomponents':'Country','Unnamed:2':'Human_Development_Index','Unnamed:4':'Life_expectancy_at_birth',
                          'Unnamed:6':'Expected_years_of_schooling','Unnamed:8':'Mean_years_of_schooling','Unnamed:10':'Gross_national_income_per_capita'}
#Se asigna la funcion ".rename" para la modificacion de indices.
indice_desarrollo.rename(columns=newIndiceDesarrolloNames,inplace=True)
#Se actualiza el DataFrame.
indice_desarrollo
```

```python
#Eliminamos las filas cuya posicion altera el correcto entendimiento y sintesis de datos del DataFrame.
#Para ello usamos la funcion ".iloc" para eliminarlos.
indice_desarrollo=indice_desarrollo.iloc[7:201]
#Se actualiza el DataFrame.
indice_desarrollo
```

```python
#Reordenamos los indices para tener una mejor organizacion de los datos.
indice_desarrollo = indice_desarrollo.reset_index(drop=True)
indice_desarrollo.head()
```

```python
#Tambien eliminamos las categorias que estan adjuntadas de forma incorrecta en el DataFrame.
filas_a_eliminar=['HIGH HUMAN DEVELOPMENT','MEDIUM HUMAN DEVELOPMENT','LOW HUMAN DEVELOPMENT']
indice_desarrollo = indice_desarrollo[~indice_desarrollo['Country'].isin(filas_a_eliminar)]
#Actualizamos el DataFrame
indice_desarrollo
```

Se da a lugar el formateo de los datos el cual inicia con el redondeo de los valores de la columna segun lo solicitado.

```python
#Asignamos a una variable las columnas para redondear.
columnas_a_redondear=['Life_expectancy_at_birth','Expected_years_of_schooling','Mean_years_of_schooling','Gross_national_income_per_capita']
#Usamos la funcion "pd.to_numeric" para cambiar el tipo de dato analizado y posterior a ello tambien la funcion "coerce".
for col in columnas_a_redondear:
  indice_desarrollo[col]=pd.to_numeric(indice_desarrollo[col], errors='coerce')
#En esta variable asignamos la cantidad de cifras a las cuales queremos redondear cada dato de su respectiva columna.
indice_desarrollo=indice_desarrollo.round({'Life_expectancy_at_birth':1,'Expected_years_of_schooling':1,'Mean_years_of_schooling':1,'Gross_national_income_per_capita':2})
#Actualizamos el DataFrame
indice_desarrollo
```

```python
#Pedimos informacion de los objetos del DataFrame para corroborar si el cambio se dio de forma efectiva.
#Lo cual se verfica tambien con el proceso de redondeo ya que no se podria hacer si sigue siendo un objeto.
indice_desarrollo.info()
```

```python
#Por ultimo asignamos la categoria de cada pais.
#Haciendo una revision de la Data original establecemos los limites y las categorias a designarse deacuerdo a dichos valores.
indice_desarrollo['category']=pd.cut(indice_desarrollo.Human_Development_Index,
                              bins=[0.3,0.5495,0.7,0.799,1],
                              labels=['LOW HUMAN DEVELOPMENT','MEDIUM HUMAN DEVELOPMENT','HIGH HUMAN DEVELOPMENT','VERY HIGH HUMAN DEVELOPMENT'],
                              ordered=True)
indice_desarrollo
```
